```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def main():
    df = pd.read_csv('./eval/evaluation_results.csv', index_col=False)

    priming_values = sorted(df['priming_value'].unique())
    demo_names = sorted(df['demo_name'].unique())

    Np = len(priming_values)
    Nd = len(demo_names)

    x = np.arange(Np)  
    bar_width = 0.8 / Nd 

    plt.figure(figsize=(12, 6), dpi=300)
    colors = plt.cm.tab10(np.linspace(0, 1, Nd)) 

    for i, demo in enumerate(demo_names):
        demo_data = df[df['demo_name'] == demo]

        means = demo_data.groupby('priming_value')['violations'].mean().reindex(priming_values)
        stds = demo_data.groupby('priming_value')['violations'].sem().reindex(priming_values)

        plt.bar(
            x + i * bar_width,
            means,
            width=bar_width,
            label=f"{demo}",
            color=colors[i],
            edgecolor="black",
            linewidth=0.7
        )

        plt.errorbar(
            x + i * bar_width,
            means,
            yerr=stds,
            fmt='none',
            ecolor='black',
            elinewidth=1.2,
            capsize=4
        )

    plt.xticks(x + bar_width * (Nd - 1) / 2, priming_values, fontsize=12)
    plt.xlabel("Priming Value", fontsize=14, labelpad=10)
    plt.ylabel("Average Norm Violations", fontsize=14, labelpad=10)
    plt.title("Norm Violations by Priming Value and Demonstration", fontsize=16, pad=15)

    plt.ylim(0, 50)

    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.legend(title="Demonstration", fontsize=11, title_fontsize=12)

    plt.tight_layout()
    plt.savefig("Violations_pub_ready.png", dpi=300)
    plt.show()
main()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def main():
    df = pd.read_csv('./eval/evaluation_results.csv', index_col=False)

    df['success'] = df['success'].astype(int)

    priming_values = sorted(df['priming_value'].unique())
    demo_names = sorted(df['demo_name'].unique())

    Np = len(priming_values)   
    Nd = len(demo_names)       

    x = np.arange(Np)
    bar_width = 0.8 / Nd

    plt.figure(figsize=(12, 6), dpi=300)
    colors = plt.cm.tab10(np.linspace(0, 1, Nd))

    for i, demo in enumerate(demo_names):
        demo_data = df[df['demo_name'] == demo]

        successs = (demo_data.groupby('priming_value')['success'].mean().reindex(priming_values) * 100)

        stds = (demo_data.groupby('priming_value')['success'].sem().reindex(priming_values) * 100)

        plt.bar(
            x + i * bar_width,
            successs,
            width=bar_width,
            label=f"{demo}",
            color=colors[i],
            edgecolor="black",
            linewidth=0.7
        )

        plt.errorbar(
            x + i * bar_width,
            successs,
            yerr=stds,
            fmt='none',
            ecolor='black',
            capsize=4,
            elinewidth=1.1
        )

    plt.xticks(x + bar_width * (Nd - 1) / 2, priming_values, fontsize=12)
    plt.xlabel("Priming Value", fontsize=14)
    plt.ylabel("Success Rate (%)", fontsize=14)
    plt.title("Success Rate by Priming Value and Demonstration Set", fontsize=16)

    plt.ylim(0, 100)
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.legend(title="Demonstration", fontsize=11, title_fontsize=12)

    plt.tight_layout()
    plt.savefig("SuccessRate_pub_ready.png", dpi=300)
    plt.show()


if __name__ == "__main__":
    main()
```
